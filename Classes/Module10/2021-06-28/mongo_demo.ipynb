```python
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.craigslist_db
collection = db.items
```

Collection: grouping of MongoDB documents. A collection is the equivalent of an RDBMS table. A collections exists within a single database. Collections do not enforce a schema. Documents within a collection can have different fields.

```python
url = "https://newjersey.craigslist.org/search/sss?sort=rel&query=guitar"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```
