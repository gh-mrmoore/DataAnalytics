```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
import sklearn as skl
import tensorflow as tf
from sklearn.datasets import make_moons

# Creating dummy nonlinear data
X_moons, y_moons = make_moons(n_samples=1000, noise=0.08, random_state=78)

# Transforming y_moons to a vertical vector
y_moons = y_moons.reshape(-1, 1)

# Creating a DataFrame to plot the nonlinear dummy data
df_moons = pd.DataFrame(X_moons, columns=["Feature 1", "Feature 2"])
df_moons["Target"] = y_moons

# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_moons, y_moons, random_state=78)

# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a Keras Sequential model and add more than one Dense hidden layer
ks_model = tf.keras.models.Sequential()

# Add dense layer(s), input layer
ks_model.add(tf.keras.layers.Dense(units=3, activation="relu", input_dim=2))

# Add input layer with activation function
ks_model.add(tf.keras.layers.Dense(units=3, activation="relu"))

# Add output layer
ks_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
ks_model.summary()
```

```python
# Compile the model and train over more than 100 epochs
ks_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

fit_model = ks_model.fit(X_train_scaled, y_train, epochs=101)
```

```python
# Evaluate the performance of model using the loss and predictive accuracy of the model on the test dataset.
model_loss, model_accuracy = ks_model.evaluate(X_test_scaled, y_test, verbose=2)
print(model_loss, model_accuracy)
```
