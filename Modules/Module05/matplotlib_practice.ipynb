```python
%matplotlib inline

# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create figure using ax.plt() - an object-oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# A different way to use ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create chart for Boston with additional information
plt.plot(x_axis, y_axis, label='Boston')
# Labels for x-axis and y-axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set y-limit between 0 and 45
plt.ylim(0, 45)
# Add title
plt.title("PyBer Fare by Month")
# Add legend
plt.legend()
```

```python
# Re-create chart with additional formatting
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Label x-axis and y-axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y-limit from 0 to 45.
plt.ylim(0, 45)
plt.xlim("Jan", "June")
# Add title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Skill Drill
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.grid()
```

```python
# Set the x-axis to a list of strings for each month.
bar_x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
bar_y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create bar plot
plt.bar(bar_x_axis, bar_y_axis)
```

```python
# Annotate bar chart
plt.bar(bar_x_axis, bar_y_axis, color="yellow", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Add title
plt.title("PyBer Fare by Month")
plt.legend()
```
